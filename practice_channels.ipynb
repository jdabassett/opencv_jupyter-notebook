```python
# import tools

import cv2
from collections import defaultdict, Counter
import matplotlib.pyplot as plt
import numpy as np 
```

```python
# load battery image

btr = cv2.imread('data/images/sample_image.png')

btr_rgb = cv2.cvtColor(btr,cv2.COLOR_BGR2RGB)
btr_gry = cv2.cvtColor(btr,cv2.COLOR_BGR2GRAY)

wb,hb,cb = btr.shape
btr_blank = np.zeros((wb,hb,cb),np.uint8) 
btr_blank1 = np.zeros_like(btr,np.uint8) # same as above
```

```python
# calculate histogram

plt.figure()
plt.title('Colour Histogram')
plt.xlabel('Bins')
plt.ylabel('# of pixels')
colors_name = ["red", "green", "blue"]
colors_data = [] 

for idx, color in enumerate(colors_name):
    hist = cv2.calcHist([btr_rgb],[idx],None,[256],[0,255]).astype('int')
    plt.plot(hist,color=color)
    plt.xlim([0,256])
    colors_data.append(hist)
    idx += 1
    
plt.show()

# print(colors_data)
```
