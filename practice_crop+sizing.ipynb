```python
# import tools

import cv2
from collections import defaultdict, Counter
import matplotlib.pyplot as plt
import numpy as np 
```

```python
# load battery image

btr = cv2.imread('data/images/sample_image.png')

btr_rgb = cv2.cvtColor(btr,cv2.COLOR_BGR2RGB)
btr_gry = cv2.cvtColor(btr,cv2.COLOR_BGR2GRAY)

wb,hb,cb = btr.shape
btr_blank = np.zeros((wb,hb,cb),np.uint8) 
btr_blank1 = np.zeros_like(btr,np.uint8) # same as above
```

```python
# resize images

btr_dbl = cv2.resize(btr_rgb, None, fx=2, fy=2, interpolation=cv2.INTER_LANCZOS4)
btr_hlf = cv2.resize(btr_rgb, None, fx=0.2, fy=0.2, interpolation = cv2.INTER_CUBIC)
btr_500 = cv2.resize(btr_rgb, (500, 500), interpolation=cv2.INTER_CUBIC)


plt.figure(figsize=(10,10))
plt.subplot(1,4,1)
plt.title("Original")
plt.imshow(btr_rgb)

plt.subplot(1,4,2)
plt.title("Half")
plt.imshow(btr_hlf)

plt.subplot(1,4,3)
plt.title("Double")
plt.imshow(btr_dbl)

plt.subplot(1,4,4)
plt.title("500x500")
plt.imshow(btr_500)

print(btr_rgb.shape, btr_dbl.shape, btr_hlf.shape, btr_500.shape)
```

```python
# flip image

btr_flpy = cv2.flip(btr_rgb,0)
btr_flpx = cv2.flip(btr_rgb,1)
btr_flpxy = cv2.flip(btr_rgb,-1)

plt.figure(figsize=(10,10))
plt.subplot(1,4,1)
plt.title("Original")
plt.imshow(btr_rgb)

plt.subplot(1,4,2)
plt.title("Flip Y")
plt.imshow(btr_flpy)

plt.subplot(1,4,3)
plt.title("Flip X")
plt.imshow(btr_flpx)

plt.subplot(1,4,4)
plt.title("Flipn XY")
plt.imshow(btr_flpxy)
```

```python
# crop and slice images

rows = hb//2
cols = wb//2

btr_slc = btr_rgb[0:rows,0:cols]

plt.figure(figsize=(20,20))
plt.subplot(1,2,1)
plt.title("Original")
plt.imshow(btr_rgb)

plt.subplot(1,2,2)
plt.title("Sliced")
plt.imshow(btr_slc)
```
